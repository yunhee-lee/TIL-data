```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("Sample Data/sales.csv")
sales.head(10)
```

```python
sales_desc = sales.iloc[:, 1:].describe()
sales_desc
```

```python
xlist = ['X1', 'X2', 'X3', 'X4']
ylist = ['Y1', 'Y2', 'Y3']
exam = sales[xlist]
perform = sales[ylist]
```

```python
# 변수 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

zexam = scaler.fit_transform(exam)
zperform = scaler.fit_transform(perform)
```

```python
# 정준상관분석 실행
from sklearn.cross_decomposition import CCA
# CCA(n_components = 2, scale = True) << default
ca = CCA(n_components = 3)
ca.fit(exam, perform)
X_c, Y_c = ca.transform(exam, perform)
```

```python
# 정준상관점수 데이터프래임 만들기
cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":X_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":X_c[:, 1],
                       "CCX_3":X_c[:, 2],
                       "CCY_3":X_c[:, 2]})
cc_res.head()
```

```python
# 제1 정준상관계수
np.corrcoef(X_c[:, 0], Y_c[:, 0])[0, 1]
```
