```python
# 함수 만들기 (def)
def if_test(x) :
  if (x % 2 == 0) :
    print("짝수")
  elif (x % 2 == 1) :
    print("홀수")
  else :
    print("자연수가 아님")

if_test(3.8)
```

```python
import numpy as np
def my_sums(a=0, b=10) :
  sum1 = 0
  sum2 = 0
  data = np.arange(a, b+1)
  for i in data :
    sum1 = sum1 + i
    sum2 = sum2 + i**2
  return sum1, sum2, len(data)
```

```python
a = my_sums(0, 10)
a
```

```python
# 이윤희 202235-367728
def mywage(whour):
  g_wage = 10000
  ot_wage = g_wage * 1.5
  ot_after = 40
  if (whour <= 40) :
    wage = whour * g_wage
  elif (whour > 40) :
    wage = (ot_after * g_wage) + ((whour - ot_after) * ot_wage)
  else :
    print("error")
  return(wage)

mywage(52)
```

## 출석 수업 과제 - 공통 3 (nutrient.txt)

```python
import pandas as pd

# 데이터 읽어오기
nutrient = pd.read_csv('/content/sample_data/nutrient.txt', sep = '\s+', header = None, index_col = 0)
nutrient.head()
```

```python
# index와 column 이름 설정
nutrient.index.name="id"
nutrient.columns = ["calcium", "iron", "protein", "vitamin A", "vitamin C"]
nutrient
```

```python
nutrient_drop = nutrient[nutrient['calcium'].isin([0]) | nutrient['iron'].isin([0])
         | nutrient['protein'].isin([0]) | nutrient['vitamin A'].isin([0]) 
         | nutrient['vitamin C'].isin([0])]
nutrient_modi = nutrient.drop(index = nutrient_drop.index)
nutrient_modi
```

```python
nutrient_drop
```

```python
nutrient.isin([0]).sum()
```

```python
nutrient_modi.describe()
```

```python
import seaborn as sns
sns.boxplot(x = "variable", y = "value", data = pd.melt(nutrient_modi))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=6)
fig.set_size_inches(30, 10)

sns.boxplot(y = "calcium", data = nutrient_modi, ax = ax[0])
sns.boxplot(y = "iron", data = nutrient_modi, ax = ax[1])
sns.boxplot(y = "protein", data = nutrient_modi, ax = ax[2])
sns.boxplot(y = "vitamin A", data = nutrient_modi, ax = ax[3])
sns.boxplot(y = "vitamin A", data = nutrient_modi[nutrient_modi["vitamin A"] < 30000], ax = ax[4])
sns.boxplot(y = "vitamin C", data = nutrient_modi, ax = ax[5])
```

```python
fig, axs = plt.subplots(1, 5, figsize=(30, 10))

plt.subplot(1, 5, 1)
plt.hist(nutrient_modi["calcium"], bins = 20)

plt.subplot(1, 5, 2)
plt.hist(nutrient_modi["iron"], bins = 20)

plt.subplot(1, 5, 3)
plt.hist(nutrient_modi["protein"], bins = 20)

plt.subplot(1, 5, 4)
plt.hist(nutrient_modi["vitamin A"], bins = 20)

plt.subplot(1, 5, 5)
plt.hist(nutrient_modi["vitamin C"], bins = 20)
```
